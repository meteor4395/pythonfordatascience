# STRING INDEXING

```python
strSample = 'learning'
```

```python
strSample.index('l')   #to find the index of substring l fron learning
```

```python
strSample.index('ning')   #to find the index of substring ning fron learning
```

```python
strSample[7]             # to get substring at index 7+1 pos
```

```python
strSample[-2]            #2nd last pos substring
```

# List: Indexing
## Syntax: list_name.index(element,start,end)

```python
lstsample=[1,2,'a','sam',2]          #list sample
```

```python
lstsample.index('sam')               # to find index of element 'sam'
```

```python
lstsample[2]                         # find the element at 3rd pos
```

```python
lstsample[-1]                        # find last element in the list
```

## Array : Indexing

```python
from array import *                              #import array module
```

```python
arrSample = array('i',[1,2,3,4])                 # array with int type
```

```python
for x in arrSample: print(x)                    #print values of 'arrSample'
```

## Tuple:Indexing

```python
tupSample = (1,2,3,4,3,'py')
```

```python
tupSample.index('py')     #find pos of py
```

```python
tupSample[2]              # find element at 2
```

## Set:Indexing

```python
setSample = {'example',24,87.5,'data',4,'data'}  #set cant be indexed
```

```python
setSample_list = list(setSample)      # set converted to list for indexing
```

```python
print(setSample_list[2])            # print list index
#the order of elements in a set is not guaranteed, so the result might vary.
```

## Dictionary: Indexing
{key:value}---dictionary are indexed by keys

```python
dictSample = {1:'first','second':2,3:3,'four':'4'}
```

```python
dictSample[2]       #gives error
```

```python
dictSample[1]     # to find the value corresponds to key 1
```

```python
dictSample['second']
```

## Range: Indexing

```python
rangeSample = range(1,12,4)           #create and print range values
for x in rangeSample: print(x)
```

```python
rangeSample.index(0)                  # value error: 0 not in range
```

```python
rangeSample.index(9)                   # to find index of element 9
```

```python
rangeSample[1]                          #to find element at index 1
```
